```python
#x以下のものがK個以上あるか？→二分探索
N,K = map(int,input().split())
A = list(map(int,input().split()))
AA = [A[i]**2 for i in range(N) if A[i] != 0]
AA.sort()
PA = [A[i] for i in range(N) if A[i] > 0] #非負
MA = [-A[i] for i in range(N) if A[i] < 0] #負
lpa = len(PA)
lma = len(MA)
lzero = N - lpa - lma
ll = lpa*lma + lzero*(lpa+lma) + lzero*(lzero-1)//2
MA_M = [-MA[i] for i in range(lma)]
MA_M.sort()
PA.sort()
MA.sort()
 
top = 10**18+1
bot = -10**18-1
 
from bisect import bisect
 
def simple_bisect(L,val,l):
    if l == 0:
        return 0
    if L[0]>val:
        return 0
    upper = l
    lower = 0
    while upper-lower>1:
        cen = (upper+lower)//2
        if L[cen] > val:
            upper = cen
        else:
            lower = cen
    return upper
 
while top-bot > 1:
    cen = (top+bot)//2
    if cen >= 0:
        S = 0
        for a in PA:
            S += simple_bisect(PA,cen//a,lpa)
        for a in MA:
            S += simple_bisect(MA,cen//a,lma)
        S -= bisect(AA,cen)
        S//=2
        S += ll
        #print(S)
        if S >= K:
            top = cen
        else:
            bot = cen
    else:
        S = 0
        for a in PA:
            S += simple_bisect(MA_M,cen//a,lma)
        if S >= K:
            top = cen
        else:
            bot = cen
print(top)
```
